### Task 1:
Using `FreqDist`, find the following from **textN**:

- Words that are *hapaxes* (unique) and those that occur less than 5 times
- Draw a plot showing mapping from **word length** to **word frequency**
- Find **10 most frequent words** occurring at the end of the sentence


```python
fdist_task1 = FreqDist(text6)
hapaxes = fdist_task1.hapaxes()
rare_words = [word for word, freq in fdist.items() if freq < 5]
```

```python
print("Hapaxes:", hapaxes)
```

```python
print("Words occurring less than 5 times:", rare_words)
```

```python
word_len_freq_dependence = {}

for word, freq in fdist_task1.items():
    length = len(word)
    word_len_freq_dependence[length] = word_len_freq_dependence.get(length, 0) + freq
```

```python
import matplotlib.pyplot as plt

plt.bar(word_len_freq_dependence.keys(), word_len_freq_dependence.values())
plt.xlabel("Word Length")
plt.ylabel("Frequency")
plt.title("Word Length & Word Frequency")
plt.show()
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize

sentences = sent_tokenize(" ".join(text6))

end_words = []
for sent in sentences:
    tokens = word_tokenize(sent)
    if len(tokens) > 1: 
        end_words.append(tokens[-2])

end_fdist = FreqDist(end_words)
print("10 most frequent words in the end of the sentence", end_fdist.most_common(10))
```

### Task 2:
Create custom NLTK text of your favourite (public domain) book.
Practice using `concordance`, `similar`, `dispersion_plot`, `FreqDist` on the resulting text.

```python
import nltk
f = open('cat-warriors.txt', 'r', encoding='utf-8')
raw = f.read()
tokens = nltk.word_tokenize(raw)
text = nltk.Text(tokens)
```

```python
text.concordance('fight')
```

```python
text.similar('fight')
```

```python
text.dispersion_plot(['fight', 'Twoleg', 'hungry'])
```

```python
fdist = nltk.FreqDist(text)
fdist.most_common(50)
```

```python
fdist.plot(50, cumulative=True)
```

```python
rare_words = [word for word, freq in fdist.items() if freq == 1]
print(rare_words)
```
