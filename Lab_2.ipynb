### Task 0: 
Compare frequency distributions of stemmed and unstemmed NLTK corpora. Display most commonly used stems from nltk.text corpora on a plot.

```python
import nltk
from nltk.book import text1 
from nltk.stem import PorterStemmer
import matplotlib.pyplot as plt
nltk.download('corpora/gutenberg')
nltk.download('book')


porter = PorterStemmer()

words = [word.lower() for word in text1 if word.isalpha()]
stemmed_words = [porter.stem(word) for word in words]

fdist_original = nltk.FreqDist(words)
fdist_stemmed = nltk.FreqDist(stemmed_words)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
fdist_original.plot(20, title="Top 20 Original Words")

plt.subplot(1, 2, 2)
fdist_stemmed.plot(20, title="Top 20 Stemmed Words")

plt.tight_layout()
plt.show()
```

### Task 2:
Implement Wagner-Fischer (or Vintsyuk) algorithm for string distance.

Modify the algorithm so that substitution operation cost depends on the key proximity on QWERTY keyboard.

Or consider this table directly:

Implement another modification to the algorithm: include transposition operation, so that you compute a Damerau-Levenshtein distance.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import math
```

```python
qwerty = {
    'q': (0, 0), 'w': (0, 1), 'e': (0, 2), 'r': (0, 3), 't': (0, 4), 'y': (0, 5), 'u': (0, 6), 'i': (0, 7), 'o': (0, 8), 'p': (0, 9),
    'a': (1, 0.5), 's': (1, 1.5), 'd': (1, 2.5), 'f': (1, 3.5), 'g': (1, 4.5), 'h': (1, 5.5), 'j': (1, 6.5), 'k': (1, 7.5), 'l': (1, 8.5),
    'z': (2, 1), 'x': (2, 2), 'c': (2, 3), 'v': (2, 4), 'b': (2, 5), 'n': (2, 6), 'm': (2, 7)
}

def qwerty_distance(a, b):
    if a not in qwerty or b not in qwerty:
        return 2
    ax, ay = qwerty[a]
    bx, by = qwerty[b]
    return math.sqrt((ax - bx) ** 2 + (ay - by) ** 2)

def substitution_cost(a, b):
    if a not in qwerty or b not in qwerty:
        return 2.0
    return 0 if a == b else 2.0

def qwerty_cost(a, b):
    return 0 if a == b else qwerty_distance(a, b)
```

### Wagner-Fischer

```python
def wagner_fischer(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n+1) for _ in range(m+1)]
    for i in range(m+1): dp[i][0] = i
    for j in range(n+1): dp[0][j] = j

    for i in range(1, m+1):
        for j in range(1, n+1):
            cost = substitution_cost(s1[i-1], s2[j-1])
            dp[i][j] = min(
                dp[i-1][j] + 1,
                dp[i][j-1] + 1,
                dp[i-1][j-1] + cost
            )
    return dp[m][n]
```

### Wagner-Fischer(qwerty)

```python
def wagner_fischer_qwerty(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0.0] * (n+1) for _ in range(m+1)]
    for i in range(m+1): dp[i][0] = i
    for j in range(n+1): dp[0][j] = j

    for i in range(1, m+1):
        for j in range(1, n+1):
            cost = qwerty_cost(s1[i-1], s2[j-1])
            dp[i][j] = min(
                dp[i-1][j] + 1,
                dp[i][j-1] + 1,
                dp[i-1][j-1] + cost
            )
    return dp[m][n]
```

### Damerau Levenshtein

```python
def damerau_levenshtein(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n+1) for _ in range(m+1)]
    for i in range(m+1): dp[i][0] = i
    for j in range(n+1): dp[0][j] = j

    for i in range(1, m+1):
        for j in range(1, n+1):
            cost = substitution_cost(s1[i-1], s2[j-1])
            dp[i][j] = min(
                dp[i-1][j] + 1,
                dp[i][j-1] + 1,
                dp[i-1][j-1] + cost
            )
            if i > 1 and j > 1 and s1[i-1] == s2[j-2] and s1[i-2] == s2[j-1]:
                dp[i][j] = min(dp[i][j], dp[i-2][j-2] + 1)
    return dp[m][n]
```

### Damerau Levenshtein(qwerty)

```python
def damerau_levenshtein_qwerty(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0.0] * (n+1) for _ in range(m+1)]
    for i in range(m+1): dp[i][0] = i
    for j in range(n+1): dp[0][j] = j

    for i in range(1, m+1):
        for j in range(1, n+1):
            cost = qwerty_cost(s1[i-1], s2[j-1])
            dp[i][j] = min(
                dp[i-1][j] + 1,
                dp[i][j-1] + 1,
                dp[i-1][j-1] + cost
            )
            if i > 1 and j > 1 and s1[i-1] == s2[j-2] and s1[i-2] == s2[j-1]:
                dp[i][j] = min(dp[i][j], dp[i-2][j-2] + 1)
    return dp[m][n]
```

### Graphics

```python
test_words_pairs = [
    ("cat", "cut"), ("cat", "bat"), ("cat", "dog"), ("cat", "qaz"),
    ("cat", "zzz"), ("kitten", "sitting"), ("play", "plaz"),
    ("keyboard", "keyboqrd"), ("example", "exampqe"),
    ("correct", "korrect"), ("hello", "yello"),
    ("function", "punctuon"), ("semantic", "srmzntid"),
    ("alphabet", "qlphqbst"), ("structure", "xzrudtyre")
]
```

```python
words_list = []
for s1, s2 in test_words_pairs:
    words_list.append({
        "Word 1": s1,
        "Word 2": s2,
        "Standard WF": wagner_fischer(s1, s2),
        "QWERTY WF": wagner_fischer_qwerty(s1, s2),
        "Standard DL": damerau_levenshtein(s1, s2),
        "QWERTY DL": damerau_levenshtein_qwerty(s1, s2),
    })

df = pd.DataFrame(words_list)
df_melted = df.melt(id_vars=["Word 1", "Word 2"], var_name="Method", value_name="Distance")
df_melted["Pair"] = df_melted["Word 1"] + "  " + df_melted["Word 2"]
```

```python
df.sort_values(by="Standard WF", ascending=False)
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

fig, axes = plt.subplots(2, 2, figsize=(18, 10))
fig.suptitle("Edit Distance Variants", fontsize=18)

methods = ["Standard WF", "QWERTY WF", "Standard DL", "QWERTY DL"]
axes = axes.flatten()

for ax, method in zip(axes, methods):
    subset = df_melted[df_melted["Method"] == method]
    sns.barplot(data=subset, x="Pair", y="Distance", palette="coolwarm", ax=ax)

    ax.set_title(method)
    ax.set_ylim(0, 9)
    ax.tick_params(axis='x', rotation=45)
    ax.grid(True, linestyle="-", alpha=0.5)
    ax.set_axisbelow(True)
    ax.set_ylabel("Distance")

fig.tight_layout(rect=[0, 0, 1, 1])  
plt.show()
```
